```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
LOG_FILE = 'logs/log-20250520-114904.csv'

log = open(LOG_FILE, 'r')

# Read the log file into a DataFrame
df = pd.read_csv(log)

# Convert the timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Set the timestamp as the index
df.set_index('timestamp', inplace=True)

# Create a plot for each metric and save it
metrics = ['workload', 'pressure', 'queue_length', 'utilization', 'action', 'theoretical_utilization', 'response_time', 'service_time']
```

```python
df[[ 'workload']].plot(figsize=(12, 6))
plt.title('Service Time and Response Time Over Time')
plt.ylabel('Time')
plt.xlabel('Timestamp')
plt.legend()
plt.show()
```

```python
df[['utilization', 'theoretical_utilization']].plot(figsize=(12, 6))
plt.title('Utilization vs Theoretical Utilization Over Time')
plt.ylabel('Utilization')
plt.xlabel('Timestamp')
plt.legend()
plt.show()
```

```python
df_plot = df.copy()
df_plot['queue_length'] = df_plot['queue_length'].clip(upper=100)
df_plot['theoretical_utilization'] = df_plot['theoretical_utilization'] * 100
df_plot[['workload', 'queue_length', 'theoretical_utilization', 'pressure']].plot(figsize=(12, 6))
plt.title('Workload and Pressure Over Time')
plt.ylabel('Value')
plt.xlabel('Timestamp')
plt.legend()
plt.show()
```
